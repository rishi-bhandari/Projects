## Face detection using OpenCV

Thus jupyter notebook will demonstrate how to build a simple detector using python. We must not confuse face detection with face recogntion as the later one requires more learning or information to give the desired output.

Face detection is a like telling the object passing by is a car, whereas face recognizer is like to be able to tell the model of the car passing by.

```python
# After installation is complete you can import it to our program

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

OpenCV already contains pretrained classifiers, like face, eyes, smile and etc. All the classifiers are stored in an XML format. Since we are using face detection model, if not already downloaded, we can use the XML file from https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml.

```python
# face_cascade = cv2.CascadeClassifier('C:\\Users\\rishi\\Documents\\GitHub\\Projects\\Face Dectection Using Python\\face_detector.xml')

face_cascade = cv2.CascadeClassifier('face_detection.xml')
```

Next step is to input an image. Make sure there is atleast one face in the image, so we test our code.

```python
img = cv2.imread('C:\\Users\\rishi\\Documents\\GitHub\\Projects\\Face Dectection Using Python\\rishi.jpg')
```

Now comes the main part, converting the image into grayscale and using face detection code to detect faces in an image.
OpenCV loads an image in BGR format, so we need to convert it into RBG format to be able to display it in coloured format as an output.

```python
#Converting to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
plt.imshow(gray_img, cmap='gray')
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
faces = face_cascade.detectMultiScale(gray_img, 1.3, 5)

# Let us print the no. of faces found
print('Faces found: ', len(faces))
```

In the above code we can see three paramters being used by detectMultiScale function. They are input image, scaleFactor and minNeighbors respectively. The two later ones provides better performance and optimised results. 

<bold> ScaleFactor </bold> parameter determines a trade off between detection accuracy and speed. The detection window starts out at the size (min_size), and after testing all the windows of that size, the window is scaled up by scaleFactor and re-tested. We continue to proceed till the window reaches maxsize. Larger the scaleFactor, fewer the steps and faster the process, but we might miss the object whose size is in between the two tested scales. 

<bold> minNeighbor </bold> parameter is used during face detection when the  resizing of sliding window gives alot of False positives. In order to avoid the FP and get proper face rectangles, neighbothood approach is applied where if we have other rectangles in the neighborhood, then it would be okay to proceed. So minNeighbor gives us the minimum neighbouring rectangles required to consider it as a face rectangle.

```python
# To draw rectangles around the face.

for (x,y,w,h) in faces:
     cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
# (255, 0, 0) is the color of the rectangle
# 2 is the thickness of the line
```

Now we will export the result as an image file. 

```python
#convert image to RGB and show image

plt.imshow(convertToRGB(img))
```

```python
cv2.imwrite("face_detected.jpg", img)
print("Successfully saved")
```
